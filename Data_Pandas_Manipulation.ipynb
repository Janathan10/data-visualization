```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = ('C:/Users/User/OneDrive/Documentos/Python/Videogame_Sales.csv ')
data = pd.read_csv(data)
data.head(200)
```

```python
Platform_data = data.Platform.value_counts()
Genre_data = data.Genre.value_counts()
Genre_data.plot(figsize= (15,8), color =['r'], legend=True  )
Platform_data.plot(kind ='bar', figsize= (15,8), color =['gray'],legend=True)
plt.title("Platform Sales", fontsize = 30 )
plt.xticks(rotation = 0, fontsize = 10)
plt.show()
```

```python
plt.figure(figsize= (10,6))
#data.NA_Sales.plot(kind = 'hist'  , color = 'r',legend = True)
#data.EU_Sales.plot(kind = 'hist' , color = 'y', legend = True)
#data.JP_Sales.plot(kind = 'hist',  color = 'g', legend = True)
#data.Other_Sales.plot( kind = 'hist' ,color = 'b', legend = True)
plt.plot(data.Year, data.NA_Sales, 'bo')
#plt.plot(data.Year, data.EU_Sales,'ro')
#plt.plot(data.Year, data.JP_Sales,'o')
plt.title("Sales per Year on NA", fontsize =30)
plt.xticks(rotation = 0, fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('Year', fontsize = 30)
plt.ylabel('Amount', fontsize = 28)
plt.show()
plt.figure()
```

```python
plt.figure(figsize= (10,6))
plt.plot(data.Year, data.EU_Sales,'ro')
#plt.plot(data.Year, data.JP_Sales,'o')
plt.title("Sales per Year on EU", fontsize =30)
plt.xticks(rotation = 0, fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('Year', fontsize = 30)
plt.ylabel('Amount', fontsize = 28)
plt.show()
plt.figure()
```

```python
plt.figure(figsize= (20,12))
plt.plot(data.Year, data.EU_Sales,'ro', label = 'EU_Sales' )
plt.plot(data.Year, data.JP_Sales,'bo',  label = 'JP_Sales' )
plt.plot(data.Year, data.Other_Sales,'go',  label = 'Other_Sales' )
plt.plot(data.Year, data.NA_Sales,'yo',  label = 'NA_Sales' )
plt.title("Sales per Country", fontsize =40)
plt.xticks(rotation = 0, fontsize = 25)
plt.yticks(fontsize = 25)
plt.xlabel('Year', fontsize = 40)
plt.ylabel('Amount', fontsize = 38)
plt.legend()
plt.show()
```
