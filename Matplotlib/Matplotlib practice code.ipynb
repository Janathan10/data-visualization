```python
# The matplotlib library working on this code.
%matplotlib qt
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
# a pie chart is a circle
# that is divided radially depending on the data. Imagine an apple pie or a pizza cut into
# slices. A pie chart fits that description well; however, unlike pizza or pies, which are
# usually divided symmetrically, a pie chart is not necessarily radially symmetrical. It all
# depends on the data to be visualized.
# Let’s create the data to be visualized, as follows:
data = np.array([35, 25, 25, 15])
# Let’s visualize the data with a simple pie chart as follows:
plt.pie(data)
plt.show()
```

```python
# Let’s add labels as follows:
%matplotlib inline
data = np.array([35, 25, 25, 15])
mylabels = ['A', 'B', 'C', 'D']
plt.pie(data,labels = mylabels)
plt.show()
```

```python
# You can even separate the parts of the pie a bit, as follows:
data = np.array([35, 25, 25, 15])
mylabels = ['A', 'B', 'C', 'D']
explode = [0.0, 0.05, 0.1, 0.15]
plt.pie(data,labels = mylabels,explode = explode)
plt.show()
```

```python
# You can also enable shadows as follows:
data = np.array([35, 25, 25, 15])
mylabels = ['A', 'B', 'C', 'D']
explode = [0.0, 0.05, 0.1, 0.15]
plt.pie(data,labels = mylabels,explode = explode,shadow = True)
plt.show()
```

```python
# You can also add a legend to the output as follows:
data = np.array([35, 25, 25, 15])
mylabels = ['A', 'B', 'C', 'D']
explode = [0.0, 0.05, 0.1, 0.15]
plt.pie(data,labels = mylabels,explode = explode,shadow = True)
plt.legend()
plt.show()
```

```python
# You can add a title for the legend as follows:
data = np.array([35, 25, 25, 15])
mylabels = ['A', 'B', 'C', 'D']
explode = [0.0, 0.05, 0.1, 0.15]
plt.pie(data,labels = mylabels,explode = explode,shadow = True)
plt.legend(title='Data :')
plt.show()
```

```python
# Polar Charts
# You can also create polar graphs that are in the shape of pie charts. However, a
# fundamental difference from the Cartesian (X-Y) coordinate system is that in a polar
# chart the coordinate system is radially arranged, so you need the angle (theta) and
# distance from the origin (r is the radius) to visualize a point or set of points. Let’s create a
# dataset as follows:
N = 20
theta = np.linspace(0.0, 2 * np.pi, N)
r = 10 * np.random.rand(N)
# The set of points can be visualized as follows:
plt.subplot(projection='polar')
plt.bar(theta, r, bottom=0.0,
color=['r', 'g', 'b'], alpha=0.2)
plt.show()
```

```python
r = np.arange(0, 5, 0.2)
theta = 2 * np.pi * r
plt.subplot(projection='polar')
plt.plot(theta, r)
plt.show()
# This creates a simple linear visualization on a polar graph. As this is a polar graph,
# you will see a spiral-like structure
```

```python
# Let’s see a couple of examples of scatter plots on a polar graph. To start, prepare the
# data as shown here:
N = 150
r = np.random.rand(N)
theta = 2 * np.pi * np.random.rand(N)
size = r * 100
# You can visualize this as follows:
plt.subplot(projection='polar')
plt.scatter(theta, r, c=theta,s=size, cmap='hsv',alpha=0.5)
plt.show()
```

```python
# You can also show part of the visualization by setting the start and end angles, as
# follows:
%matplotlib inline

fig = plt.figure()
ax = fig.add_subplot(projection='polar')
c = ax.scatter(theta, r, c=theta, s=size, cmap='hsv', alpha=0.5)
ax.set_thetamin(0)
ax.set_thetamax(90)
plt.show()
```

```python
# 3D Contours
# You can create 3D contours with the functions contour() and contour3D().
# Let’s create some data to be visualized
fig = plt.figure()
ax = plt.axes(projection='3d')
x = np.linspace(-10, 10, 30)
y = np.linspace(-10, 10, 30)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X ** 2 + Y ** 2))
# You can create a contour as follows:
ax = fig.add_subplot(projection='3d')
ax.contour(X, Y, Z, 50, cmap='coolwarm')
plt.show()
```

```python
fig = plt.figure(figsize=(7,6))
ax = fig.add_subplot(projection='3d')
# Let’s define colors for the bars.
colors = ['r', 'g', 'b', 'c', 'm', 'y','k']
yticks = [0, 1, 2, 3, 4, 5, 6]
# Now, let’s create bar graphs with the defined colors with the
# following loop:
for c, k in zip(colors, yticks):
    x = np.arange(25)
    y = np.random.rand(25)
    ax.bar(x, y, zs=k, zdir='y',color=c, alpha=0.8)
plt.show()
```

```python
# You can also create a 3D bar graph with Matplotlib. Let’s create
# the data first, as shown here:
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(projection='3d')
x = np.arange(10) * np.arange(10)
y = np.arange(10) * np.arange(10)
x, y = np.meshgrid(x, y)
x, y = x.ravel(), y.ravel()
top = x + y
bottom = np.zeros_like(top)
width = depth = 5
# You can then show this as 3D bars as follows:
ax.bar3d(x, y, bottom, width,depth, top,shade=True,color='g')
plt.show()
```

```python
# Quiver and Stem Plots
# A quiver plot is used to represent directional entities
# (for example, vectors). Let’s define the data, as shown here:
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
x = y = z = np.arange(-0.1, 1, 0.2)
X, Y, Z = np.meshgrid(x, y, z)
u = np.cos(np.pi * X) * np.sin(np.pi * Y) * np.sin(np.pi * Z)
v = -np.sin(np.pi * X) * np.cos(np.pi * Y) * np.sin(np.pi * Z)
w = np.sin(np.pi * X) * np.sin(np.pi * Y) * np.cos(np.pi * Z)
# Finally, you can visualize the data as follows:
ax.quiver(X, Y, Z, u, v, w,length=0.1,normalize=True)
plt.show()
```

```python
# You can also create stem plots where perpendicular lines are drawn
# in the visualization. Let’s use trigonometric functions to define the
# data, as shown here
# generating some data
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-10,10,100);
y = np.sin(x);
z = np.cos(x);
fig = plt.figure()
ax = Axes3D(fig)
# plotting the stems
for i in range(len(x)):
    ax.plot([x[i], x[i]], [y[i], y[i]], [0, z[i]],
        '--', linewidth=2, color='b', alpha=.5)
# plotting a circle on the top of each stem
ax.plot(x, y, z, 'o', markersize=8,
        markerfacecolor='none', color='b',label='ib')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

```python
# 3D Volumes
# You can show 3D volumetric data as enclosed surfaces.
# Let’s create such data as follows:
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = 10 * np.outer(np.cos(u), np.sin(v))
y = 10 * np.outer(np.sin(u), np.sin(v))
z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))
# You can show this data as a sphere as follows:
ax.plot_surface(x, y, z)
plt.show()
```

```python
# You can also use the function voxels() to visualize a volume as follows:
%matplotlib qt
%matplotlib inline

ma = np.random.randint(1, 3, size=(3, 3, 3))
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.voxels(ma, edgecolor='k')
plt.show()
```
